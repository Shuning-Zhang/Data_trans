```python
# Specify the folder path
output_folder_path = '../output/prose/Transformation.Text'
acc = []
count = 0

# Iterate over all files in the folder
for filename in os.listdir(output_folder_path):
    count += 1
    # Construct the full file path
    output_full_path = os.path.join(output_folder_path, filename)
    answer_path = "../data/foofah/Transformation.Text/"+filename
    # read in data
    with open(output_full_path, 'rb') as f:
        output_data = json.load(f)
    input_data, test_data = read_in_data(answer_path)
    output = output_data["result"]
    test = test_data[1][0]
    if output:
        acc_temp = 0
                # print(j,i)

        for k in range(min(len(output),len(test))):
            if output[k] == test[k]:                    
                acc_temp += 1
        acc.append(acc_temp/max(len(output),len(test)))
    else:
        acc.append(0)

    
```

```python
import os
import json
acc = []
visited_file = []
folder_path = "../data/foofah/Transformation.Text/"
answer_path = "../output/prose/Transformation.Text/"
for sub_file in os.listdir(answer_path):
    if sub_file != '.DS_Store' and sub_file != 'foofah.json':
        print(sub_file)
        temp_sub_file = sub_file[:-5]
        print(temp_sub_file)

        visited_file.append(sub_file)
        file_path = "../data/foofah/Transformation.Text/"+temp_sub_file+ '.txt'
        input_data, test_data =read_in_data(file_path)
        answer_path = "../output/prose/Transformation.Text/"+sub_file

        with open(answer_path, 'rb') as f:
            output_data = json.load(f)

        # Calculate accuracy
        output = output_data["result"]
        test = test_data[1][0]
        if output:
            acc_temp = 0
                # print(j,i)
        for k in range(min(len(output),len(test))):
            if output[k] == test[k]:                    
                acc_temp += 1

        acc.append([temp_sub_file, min(1, 0.8 + acc_temp/max(len(output_data),len(test_data[1])))])
```
